```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('../result/result20y00.700000.txt')
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
